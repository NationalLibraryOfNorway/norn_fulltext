```python
import pandas as pd
from glob import glob
import dhlab as dh
```

```python
# Control that metadata and texts matches for each corpus

metadata_files = glob("metadata/*")

korpus_folders = glob("texts/*")

text_dct = {}
for folder in korpus_folders:
    text_dct[folder.split("/")[-1]] = glob(folder + "/*.txt")
```

```python
text_df = pd.DataFrame.from_dict(text_dct, orient="index")
text_df = text_df.stack().reset_index(level=0)
text_df.columns = ["korpus", "text_path"]
text_df['text'] = text_df['text_path'].apply(lambda x: open(x).read())
text_df["dhlabid"] = text_df["text_path"].apply(lambda x: x.split("/")[-1].split(".")[0].split("_")[1])
```

```python
dct = {}
for file in metadata_files:
    dct[file.split("/")[-1]] = pd.read_excel(file, index_col=0)
    
keys = list(dct.keys())
```

```python
def check_corpus(df):
    print("Number of rows in metadata: {}".format(len(df)))
    print("Number of unique urns in metadata: {}".format(len(df["urn"].unique())))
    print("Number of unique dhlabids in metadata: {}".format(len(df["dhlabid"].unique())))
    print("Number of rows with dhlabid {}".format(len(df[df["dhlabid"].notna()])))
    print("Number of rows without dhlabid {}".format(len(df[df["dhlabid"].isna()])))
    
    # Compare with dhlab
    c = dh.Corpus()
    c.extend_from_identifiers(df["urn"])    
    print("Korpus rows: {}".format(len(c.frame)))
    
    
```

```python
check_corpus(dct[keys[2]])
```

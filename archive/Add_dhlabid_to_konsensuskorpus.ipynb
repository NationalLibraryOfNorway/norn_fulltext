```python
import pandas as pd
import dhlab as dh
```

```python
kk = pd.read_excel("metadata/konsensus-korpus.xlsx")
```

```python
kk["temp_urn"] = kk.urn.str.split("/", expand=True).iloc[:, 1]
```

```python
c = dh.Corpus()
c.extend_from_identifiers(kk.temp_urn.tolist())
```

```python
kk
```

```python
target = c.frame.loc[:, ["urn", "dhlabid"]]
```

```python
merged = kk.merge(target, left_on="temp_urn", right_on="urn", how="left")# .to_csv("metadata/konsensus-korpus.csv", index=False
```

```python
merged.loc[merged.dhlabid.isna()]
```

```python
merged.to_excel("metadata/konsensus-korpus_metadata_dhlabid.xlsx", index=False)
```
